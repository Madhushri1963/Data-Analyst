# Step 5 feature transformation

```python
from sklearn.preprocessing import LabelEncoder  #import LabelEncoder from sklearn.preprocessing library
labelencoder = LabelEncoder()  # add variable of LabelEncoder
df_telecome['gender']=labelencoder.fit_transform(df_telecome['gender'])   #to fit & transform the Churn column data 
df_telecome['PhoneService']=labelencoder.fit_transform(df_telecome['PhoneService'])
df_telecome['InternetService']=labelencoder.fit_transform(df_telecome['InternetService'])
df_telecome['Contract']=labelencoder.fit_transform(df_telecome['Contract'])
df_telecome['PaymentMethod']=labelencoder.fit_transform(df_telecome['PaymentMethod'])
df_telecome['TotalCharges']=labelencoder.fit_transform(df_telecome['TotalCharges'])
df_telecome['Churn']=labelencoder.fit_transform(df_telecome['Churn'])      #to fit & transform the Churn column data 

df_telecome   #print the dataset
```

# Train Test - method

```python
from sklearn.model_selection import train_test_split                                                        #import train_test_split from sklearn.model_selection library
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=104, test_size=0.20, shuffle=True)
```

```python
x_train.shape        #show shape of x_train
```

```python
x_test.shape         #show shape of x_test
```

```python
y_train.shape        #show shape of y_train
```

```python
y_test.shape         #show shape of y_test
```

```python
# feature scaling using normaization method

from sklearn.preprocessing import MinMaxScaler                                  #import MinMaxScaler from sklearn.preprocessing library           
minmax = MinMaxScaler().fit(x_train)                                            # Applying the normalization scaler to training data                                                        
x_train_nrm = minmax.fit_transform(x_train)                                     #Transforming the training data into normalization
#X_train_nrm = pd.DataFrame(X_train_nrm, columns=df_nifty_new.columns)          #dataset columns assign to dataframe and add column names
x_test_nrm = minmax.transform(x_test)
x_test_nrm
```

```python
# feature scaling using normaization method

from sklearn.preprocessing import MinMaxScaler                                  #import MinMaxScaler from sklearn.preprocessing library           
minmax = MinMaxScaler().fit(y_train)                                            # Applying the normalization scaler to training data                                                        
y_train_nrm = minmax.fit_transform(y_train)                                     #Transforming the training data into normalization
#X_train_nrm = pd.DataFrame(X_train_nrm, columns=df_nifty_new.columns)          #dataset columns assign to dataframe and add column names
y_test_nrm = minmax.transform(y_test)
y_test_nrm 
```

```python
# Logistic Regression Algorithm
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(random_state = 42)
logreg.fit(x_train_nrm, y_train_nrm)

# Making predictions on traning accuracy dataset
from sklearn.metrics import accuracy_score
y_pred_logreg = logreg.predict(x_train_nrm)    #x_train data
train_accuracy = accuracy_score(y_train_nrm, y_pred_logreg)     
print(train_accuracy)


# Making predictions on testing accuracy dataset
from sklearn.metrics import accuracy_score
y_test_logreg = logreg.predict(x_test_nrm)    # x_test data
test_accuracy = accuracy_score(y_test_nrm, y_test_logreg)      #y_train, y_test
print(test_accuracy)
```

```python
# K nearest neighbors Algorithm
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 24, metric = 'minkowski', p = 2)
knn.fit(x_train_nrm, y_train_nrm)

# Making predictions on traning accuracy dataset
from sklearn.metrics import accuracy_score
y_pred_knn = knn.predict(x_train_nrm)    #x_train data
train_accuracy = accuracy_score(y_train_nrm, y_pred_knn)     
print(train_accuracy)


# Making predictions on testing accuracy dataset
from sklearn.metrics import accuracy_score
y_test_knn = knn.predict(x_test_nrm)    # x_test data
test_accuracy = accuracy_score(y_test_nrm, y_test_knn)      #y_train, y_test
print(test_accuracy)
```

```python
# Support Vector Classifier Algorithm
from sklearn.svm import SVC
svc = SVC(kernel = 'linear', random_state = 42)
svc.fit(x_train_nrm, y_train_nrm)

# Making predictions on traning accuracy dataset
from sklearn.metrics import accuracy_score
y_pred_svc = svc.predict(x_train_nrm)    #x_train data
train_accuracy = accuracy_score(y_train_nrm, y_pred_svc)     
print(train_accuracy)

# Making predictions on testing accuracy dataset
from sklearn.metrics import accuracy_score
y_test_svc = svc.predict(x_test_nrm)    # x_test data
test_accuracy = accuracy_score(y_test_nrm, y_test_svc)      #y_train, y_test
print(test_accuracy)
```

```python
# Naive Bayes Algorithm
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train_nrm, y_train_nrm)


# Making predictions on traning accuracy dataset
from sklearn.metrics import accuracy_score
y_pred_nb = nb.predict(x_train_nrm)    #x_train data
train_accuracy = accuracy_score(y_train_nrm, y_pred_nb)     
print(train_accuracy)


# Making predictions on testing accuracy dataset
from sklearn.metrics import accuracy_score
y_test_nb = nb.predict(x_test_nrm)    # x_test data
test_accuracy = accuracy_score(y_test_nrm, y_test_nb)      #y_train, y_test
print(test_accuracy)
```

```python
# Decision tree Algorithm
from sklearn.tree import DecisionTreeClassifier
dectree = DecisionTreeClassifier(criterion = 'entropy', random_state = 42)
dectree.fit(x_train_nrm, y_train_nrm)


# Making predictions on traning accuracy dataset
from sklearn.metrics import accuracy_score
y_pred_dectree = dectree.predict(x_train_nrm)    #x_train data
train_accuracy = accuracy_score(y_train_nrm, y_pred_dectree)     
print(train_accuracy)


# Making predictions on testing accuracy dataset
from sklearn.metrics import accuracy_score
y_test_dectree = dectree.predict(x_test_nrm)    # x_test data
test_accuracy = accuracy_score(y_test_nrm, y_test_dectree)      #y_train, y_test
print(test_accuracy)
```

```python
# Random forest Algorithm
from sklearn.ensemble import RandomForestClassifier
ranfor = RandomForestClassifier(n_estimators = 11, criterion = 'entropy', random_state = 42)
ranfor.fit(x_train_nrm, y_train_nrm)

# Making predictions on traning accuracy dataset
from sklearn.metrics import accuracy_score
y_pred_ranfor = ranfor.predict(x_train_nrm)    #x_train data
train_accuracy = accuracy_score(y_train_nrm, y_pred_ranfor)     
print(train_accuracy)


# Making predictions on testing accuracy dataset
from sklearn.metrics import accuracy_score
y_test_ranfor = ranfor.predict(x_test_nrm)    # x_test data
test_accuracy = accuracy_score(y_test_nrm, y_test_ranfor)      #y_train, y_test
print(test_accuracy)
```

```python
# Evaluating using accuracy_score metric
from sklearn.metrics import accuracy_score
accuracy_knn = accuracy_score(y_train_nrm, y_pred_knn)
accuracy_svc = accuracy_score(y_train_nrm, y_pred_svc)
accuracy_nb = accuracy_score(y_train_nrm, y_pred_nb)
accuracy_ranfor = accuracy_score(y_train_nrm, y_pred_ranfor)
accuracy_dectree = accuracy_score(y_train_nrm, y_pred_dectree)
accuracy_logreg = accuracy_score(y_train_nrm, y_pred_logreg)
```

```python
# Accuracy on test set
print("K Nearest neighbors: " + str(accuracy_knn * 100))
```

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_train_nrm, y_pred_knn)
cm = confusion_matrix(y_train_nrm, y_pred_svc)
cm = confusion_matrix(y_train_nrm, y_pred_nb)
cm = confusion_matrix(y_train_nrm, y_pred_ranfor)
cm = confusion_matrix(y_train_nrm, y_pred_dectree)
cm = confusion_matrix(y_train_nrm, y_pred_logreg)
cm
```
